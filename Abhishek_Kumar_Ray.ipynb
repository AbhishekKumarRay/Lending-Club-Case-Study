### Finding Fields that contains High Missing Values.

```python
missingValues = loanData.isnull().sum() * 100 / len(loanData)
missingValues.to_frame(name="percent_missing")

missingValues = missingValues.reset_index()
missingValues.columns = ['field', 'percent_missing']
missingValues = missingValues.sort_values(by="percent_missing", ascending=False)


missingValues = missingValues[missingValues['percent_missing'] > 0]
figure, axix = plt.subplots(figsize =(6, 6))
 
axix.barh(missingValues["field"], missingValues["percent_missing"])

for s in ['top', 'bottom', 'left', 'right']:
    axix.spines[s].set_visible(False)
 
axix.xaxis.set_ticks_position('none')
axix.yaxis.set_ticks_position('none')

axix.set_title('Percent of missing values in per field')
axix.set_xlabel('%')


axix.xaxis.set_tick_params(pad = 20)
axix.yaxis.set_tick_params(pad = 20)
 
axix.invert_yaxis()
 
for i in axix.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 

plt.show()
```

* Based on the nature of distribution **"loan_amnt", "int_rate", "annual_inc", "installment"** can be bucketed for further analysis.

##### CREATING BUCKETS FOR BIVARIATE ANALYSIS

```python
# Creating categorical buckets for loan amount
loanData["loan_amnt_bkt"] = pd.cut(loanData['loan_amnt'], 
                               list(range(0,40001,7000)), 
                               labels = ['0 - 7k', '7k - 14k', '14k - 21k', '21k - 28k', '28k+'])

# Creating categorical buckets for interest rate
loanData["int_rate_bkt"] = pd.cut(loanData['int_rate'], 
                               list(range(5,26,5)), 
                               labels = ['5 - 10', '10 - 15', '15 - 20', '20+'])


# Creating categorical buckets for annual income
loanData['annual_inc_bkt'] = pd.cut(loanData['annual_inc'],
                                 list(range(0,250001,50000)), 
                                 labels = ['0 - 50k', '50k - 100k', '100k - 150k', '150k - 200k', '200k+'])

# Creating categorical buckets for installment
loanData["installment_bkt"] = pd.cut(loanData['installment'], 
                               list(range(0,1401,200)), 
                               labels = ['0 - 200', '200 - 400', '400 - 600', '600 - 800', 
                                         '800 - 1000', '1000 - 12000', '12000+'])
```

##### VISUALIZING THE BUCKETS

```python
# Creating plots
fig = plt.figure(figsize=(20,10))
spec = fig.add_gridspec(2,2)
spec.update(wspace=0.2, hspace=0.3)
sec_01 = fig.add_subplot(spec[0,0])
sec_02 = fig.add_subplot(spec[0,1])
sec_03 = fig.add_subplot(spec[1,0])
sec_04 = fig.add_subplot(spec[1,1])

# Adding color preference
bg_color = '#ffd9d9'
for selection in [fig, sec_01, sec_02, sec_03, sec_04]:
    selection.set_facecolor(bg_color)

# Plotting the graph
sec = [sec_01, sec_02, sec_03, sec_04]

cnt = 0
for feature in ['loan_amnt_bkt', 'int_rate_bkt', 'installment_bkt', 'annual_inc_bkt']:
    sns.countplot(loanData[feature], ax=sec[cnt])
    sec[cnt].set_ylabel('COUNT' , fontsize=12)
    cnt+=1

# Naming the axis
sec_01.set_xlabel('LOAN_AMOUNT_BUCKET' , fontsize=12)
sec_02.set_xlabel('INTEREST_RATE_BUCKET' , fontsize=12)
sec_03.set_xlabel('INSTALLMENT_BUCKET' , fontsize=12)
sec_04.set_xlabel('ANNUAL_INCOME_BUCKET' , fontsize=12)

plt.show()
```

4. Loans carrying the highest interest rates within their respective categories are more prone to charge-offs.
5. Loans designated for housing, debt consolidation, and small business, featuring the highest interest rates, are more likely to result in charge-offs.
6. Longer loan tenures tend to have higher interest rates, and individuals with terms between 15-20% are more susceptible to charge-offs.
7. Shorter loan tenures tend to have lower interest rates, and individuals with terms between 10-15% are more prone to charge-offs.
8. Borrowers falling under grade E, F, or G, with an average loan amount exceeding 15,000, are at an increased risk of charge-offs.
9. Borrowers with a verification status of "not-verified," holding loan amounts between 5,000 and 12,500, are more likely to experience charge-offs.
10. Borrowers with "not-verified" and "verified" verification statuses exhibit an equal proportion in charge-off occurrences.
11. An increase in the number of installments corresponds to a higher likelihood of charge-offs for borrowers with elevated interest rates.
